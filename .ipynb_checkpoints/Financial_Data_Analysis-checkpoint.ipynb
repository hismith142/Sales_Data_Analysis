#### Task 2: Data preprocessing

```python
#importing needed libraries
import pandas as pd

#creating initial dataframe object
df = pd.read_csv('all_data.csv')

#getting an overview of datatypes
df.info()
df.head()
```

First, I'll assess the dataframe to get an idea of how 'full' the data is (checking for NaN values).

```python
#evaluating the amount of null values in dataframe
null = df.isna().any(axis=1).sum()
print(f"Dataframe has {null} rows with missing values.")
```

Since the dataframe is over 99% full, dropping NaN rows seems to be a reasonable choice as compared to imputation which may result in false correlations.

```python
#Dropping NaN rows from dataframe
df = df.dropna(how='all')

#Revisiting number of non-full rows
null = df.isna().any(axis=1).sum()
print(f"Dataframe has {null} rows with missing values.")
```

There are rows in the 'Order Date' column that start with 'or'. I'll take them out of the dataframe.

```python
temp = df[df['Order Date'].str[0:2] != 'Or']
df = temp.copy()
```

Now we can begin to update the datatypes of the columns and finish preprocessing. First, I'll add columns 'Day', 'Month', 'Hour', and 'Minute'. Then, I will delete the 'Order Time' column as it is no longer needed. Additionally, I will not create a 'Year' column for the sake of memory efficiency as the data only spans one year (2019).

```python
#Inserting 'Month' column
df['Month'] = df['Order Date'].str[0:2].astype('int32')
df.head()
```

```python
#Inserting 'Day' column
df['Day'] = df['Order Date'].str[3:5].astype('int32')
df.head()
```

```python
#Inserting 'Hour' column
df['Hour'] = df['Order Date'].str[9:11].astype('int32')
df.head()
```

```python
#Inserting 'Minute' column
df['Minute'] = df['Order Date'].str[12:14].astype('int32')
df.head()
```

```python
#Removing 'Order Date' column
df = df.drop('Order Date', axis=1)
df.head()
```

While the exact address may be important for other scenarios, I'll only be looking at the state of the purchase address. Therefore, I'll update the dataframe by adding a 'State' column and deleting the 'Purchase Address' column.

```python
df['State'] = df['Purchase Address'].str[-9:-6].astype('string')
df = df.drop('Purchase Address', axis=1)
df.head()
```

```python
#Converting remainting row types
df['Quantity Ordered'] = df['Quantity Ordered'].astype('int32')
df['Price Each'] = df['Price Each'].astype('float64')
```

Finally, I will add a 'Order Price' column that will contain the total price of each order.

```python
#Adding 'Order Price' column
df['Order Price'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

```python
state_sales = df.groupby('State')[['Quantity Ordered', 'Order Price']].sum()
state_sales
```

Initial Analysis: California is responsible for the largest portion of the company's gross profit. The company is currently only selling to 8 states in the US.

Additional Note: There seems to be a direct correlation between quantity of sales and gross profits across all states.

In order to accurately assess the data, population must be taken into account when looking at the number of sales and gross profit per state. To assess this, I will apply a weight to each state according to their respective population as of July 2019.

```python
#dictionary to store population adjusted sale amounts
state_pop_data = {
    'CA' : 39.44 / state_sales.iloc[0]['Order Price'],
    'WA' : 7.86 / state_sales.iloc[7]['Order Price'],
    'TX' : 29.53 / state_sales.iloc[6]['Order Price'],
    'OR' : 4.24 / state_sales.iloc[5]['Order Price'],
    'NY' : 19.84 / state_sales.iloc[4]['Order Price'],
    'ME' : 1.38 / state_sales.iloc[3]['Order Price'],
    'MA' : 6.98 / state_sales.iloc[2]['Order Price'],
    'GA' : 10.8 / state_sales.iloc[1]['Order Price']
}


states = []
pop_adjusted_vals = []
for key, value in state_pop_data.items():
    states.append(key)
    pop_adjusted_vals.append(value)
    
plt.pie(pop_adjusted_vals, labels=states)
plt.title('Gross Profit by State (Adjusted for Pop.)')
plt.show()
```

Further Analysis: Now that the data has been adjusted for population, we see that Texas has taken the lead for gross profit! Maine has gained a sizeable portion of the pie chart, and California has shrunk significantly. By population metrics alone (and provided the strategy is to strengthen weaker markets), it would be best to focus marketing efforts toward Massachusetts.

## Section 2: Classification Machine Learning Model

#### Task 1: Preparing data for model training
Due to the nature of the desired model, I will be implimenting one hot encoding to turn the catagories of predictors into integer encoded binary vectors. The target column will be 'Product'. For the sake of this model I will be excluding the possibility of multiple items in a cart.

```python
#revisiting main dataframe object
df.head()
```

```python
#dropping duplicate Order IDs. (will make it such that one product is associated with one order time/location).
df = df.drop_duplicates(subset=['Order ID'])
df.head()
```

```python
#one-hot-encoding the categorical columns (time isn't on a numerical scale so it will be treated like a category)
oh_df = pd.get_dummies(df, columns = ['Month', 'Day', 'Hour', 'Minute', 'State'])
oh_df.head()
```

```python
df = oh_df.drop(['Order ID'], axis=1)
df.head()
```

```python
predictors_df = df.drop(['Product'], axis=1)
predictors_list = predictors_df.columns.values.tolist()

#target
y = df['Product']
#predictors
x = df[predictors_list]
```

```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
```

Now that all the proper splitting has been done via the pandas library, it's time to start importing from sklearn

```python
#importing classification models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

#importing performance metrics for model
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

#importing preprocessing
from sklearn.model_selection import train_test_split

#creating model pipeline
pipeline = []
pipeline.append(LogisticRegression(solver='liblinear'))
pipeline.append(DecisionTreeClassifier())
pipeline.append(RandomForestClassifier())
pipeline.append(SVC())
pipeline.append(KNeighborsClassifier())
pipeline.append(GaussianNB())

model_list = ['LogisticRegression', 'DecisionTreeClassifier', 
              'RandomForestClassifier', 'SVC', 'KNeighborsClassifier', 
              'GaussianNB']

acc_list = []
auc_list = []
cm_list = []
```

```python
#partitioning training and validation data
x_train, x_valid, y_train, y_valid = train_test_split(x, y, random_state=1, test_size = 0.25)

for model in pipeline:
    model.fit(x_train, y_train)
    pred = model.predict(x_valid)
    acc_list.append(metrics.accuracy_score(y_valid, pred)
    #fpr, ftr, _threshholds = metrics.roc_curve(y_valid, pred)
    #auc_list.append(round(metrics.auc(fpr, ftr), 2))
    #cm_list.append(confusion_matrix(y_valid, pred))

```
